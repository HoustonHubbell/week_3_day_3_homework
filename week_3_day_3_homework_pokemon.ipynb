# Goal 
### build a shopping cart program with prices and quantities using objects and a dictionary

```python
from IPython.display import clear_output as clear
# Create a class called cart that retains items and has methods to add, remove, and show

#This class is responsible for the cart object and actions of the cart

class Cart():
    def __init__(self):
        self.items = {}
    
    def add(self):
        clear()
        new_item = input("What would you like?")
        quantity = int(input(f"How many would you like? insert number"))
        if new_item not in self.items.keys():
            self.items[new_item] = quantity
        else:
            self.items[new_item] == quantity
        print("{quantity} {new_item}'s have been added")
        
    def remove(self):
        clear()
        discard = input("What would you like to toss?")
        quantity = int(input("How many would you like to toss?"))
        try:
            self.items[discard] -= quantity
            if self.items[discard] <= 0:
                del self.items[discard]
        except:
            print(f"{discard} was not in your cart")
        self.show()
        
    def show(self):
        print("Your cart has the following items and amounts")
        print("Items .......Quantity")
        for item.quantity in self.items.items():
            print({item}, {quantity})
                  
    def checkout(self):
        clear()
        if self.items:
            print("Thanks for shopping")
            self.show()
        else:
            print("Buy something next time")
#Control logic  and flow
class Main:
    def show_instructions(self):
        print("""
        Welcome to the online shopping program.
        options:
        [1] Show current cart
        [2] Add item
        [3] remove item
        [4] Quit
        """)
        
        #driver code
    def run(self):
        my_cart = Cart()
        
        while True:
            self.show_instructions()
            choice = input("What would you like to do")
            if choice == '1':
                if my_cart.items:
                    my_cart.show()
                else:
                    print('Nothing')
                    
            elif choice == '2':
                my_cart.add()
            elif choice == '3':
                if my_cart.items:
                    my_cart.remove()
                else:
                    print('cart is empty')
            elif choice == '4':
                my_cart.checkout()
                break
            else:
                print('invalid response')
                
my_program = Main()
my_program.run()
                
        
```

## Lets create a class called 'Pokemon' and create our pokemon as instances

```python
class Pokemon():
    def __init__(self,name):
        self.name = name
        self.types = []
        self.abilities = []
        self.weight = None
        self.poke_api_call()
        
    def poke_api_call(self):
        r = requests.get(f'https://pokeapi.co/api/v2/pokemon/{self.name}')
        if r.status_code == 200:
            pokemon = r.json()
        else:
            print(f"Ran into an issue {r.status_code}")
            return
        self.name = pokemon['name']
        self.types = [ poke['type']['name'] for poke in pokemon['types'] ]
        self.abilities = [ poke['ability']['name'] for poke in pokemon['abilities'] ]
        self.weight = pokemon['weight']
        print(f"{self.name}'s data has been updated!")
        
    def __repr__(self):
        return f"You caught a {self.name}"
```

If your pokemon can't evolve any further print a message that says "\<name of pokemon> can't evolve."

```python
from time import sleep

class Evolver:
    
    def evolve(self):
        
        r = requests.get(f"https://pokeapi.co/api/v2/pokemon-species/{self.name}/")
        if r.status_code == 200:
            pokemon_species = r.json()
        else:
            print(f'Ran into an issue {r.status_code}')
            return
        
        
        r = requests.get(pokemon_species['evolution_chain']['url'])
        if r.status_code == 200:
            ev_chain = r.json()
            ev_chain = ev_chain['chain']
        else:
            print(f"Ran into an issue {r.status_code}")
            return
        
        base_name = ev_chain['species']['name']
        
        evolution = ev_chain['evolves_to'][0]
        evolution_name = evolution['species']['name']
        
        if base_name == self.name:
            pass
        elif evolution_name == self.name:
            evolution_name = evolution['evolves_to'][0]['species']['name']
        else:
            print(f"You cant evolve {self.name} again")
            return
            
        print("....")
        sleep(1)
        print(f"Your {self.name} is evolving")
        self.display()
        sleep(1)
        print(".....")
        self.name = evolution_name
        self.poke_api_call()
        self.display()
        print(f"Congrats you have a new {self.name}")
```
